```python
from splinter import Browser
from bs4 import BeautifulSoup
import re
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from sqlalchemy import create_engine
from config import PGpassword
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.nytimes.com/2021/05/04/us/politics/biden-vaccination-strategy.html'
browser.visit(url)
```

```python
html = browser.html
NYT = BeautifulSoup(html, 'html.parser')
NYT
```

```python
NYTData = []
news = NYT.find_all('p')
news

for i in news: 
    text = i.text
    NYTData.append(text)
    print(text)
```
